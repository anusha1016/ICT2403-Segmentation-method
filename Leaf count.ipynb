```python
import cv2
import numpy as np

# Load image
image = cv2.imread("01.jpg")  # Change this to your actual image file
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)  # Convert to HSV color space

# Define green color range (adjust if needed)
lower_green = np.array([30, 40, 40])  # Lower bound for green
upper_green = np.array([90, 255, 255])  # Upper bound for green

# Create a binary mask for green color
mask = cv2.inRange(hsv, lower_green, upper_green)

# Apply morphological operations to remove noise
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# Find contours of leaves
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter out small contours (to remove noise)
min_leaf_size = 100  # Adjust this value based on image
leaf_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > min_leaf_size]

# Count the leaves
leaf_count = len(leaf_contours)
print(f"Leaf Count: {leaf_count}")

# Draw contours on the original image
output = image.copy()
cv2.drawContours(output, leaf_contours, -1, (0, 255, 0), 2)  # Green contours

# Show images
cv2.imshow("Green Mask", mask)
cv2.imshow("Detected Leaves", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
