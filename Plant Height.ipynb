```python
import cv2
import numpy as np

# Load the image
image = cv2.imread("01.jpg")  # Change this to your actual image file
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)  # Convert to HSV color space

# Define range for green color (adjust if needed)
lower_green = np.array([30, 40, 40])  # Lower bound for green
upper_green = np.array([90, 255, 255])  # Upper bound for green

# Create a mask for green color
mask = cv2.inRange(hsv, lower_green, upper_green)

# Apply morphological operations to remove noise
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# Find contours of the plant
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

if contours:
    # Find the tallest contour (assumed to be the plant)
    plant_contour = max(contours, key=lambda c: cv2.boundingRect(c)[3])

    # Get bounding box (x, y, width, height)
    x, y, w, h = cv2.boundingRect(plant_contour)

    # Draw bounding box on original image
    output = image.copy()
    cv2.rectangle(output, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Print the plant height
    print(f"Plant Height: {h} pixels")

    # Display the images
    cv2.imshow("Green Mask", mask)
    cv2.imshow("Detected Plant Height", output)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("No plant detected!")
```
